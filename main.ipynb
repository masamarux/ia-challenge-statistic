# Setting up!

```python
import pandas as pd
# import matplotlib.dates as mdates
import matplotlib.pyplot as plt
```

### Data and Dataframe

```python
dict_billing = {
    'date_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'value': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

```python
df_billing = pd.DataFrame.from_dict(dict_billing)
```

```python
df_billing["date_ref"] = pd.to_datetime(df_billing["date_ref"])
```

# Statistics Time!

```python
# values mean is
df_billing.value.mean()
```

```python
# I am having trouble to plot formated data, even using mathplot date formater, I'll trying something different next...
df_billing.plot.bar(x='date_ref', y='value')
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
df_billing.plot(kind='bar', x='date_ref', y='value', ax=ax, legend=False)

ax.set_xticklabels(df_billing.sort_values(by='date_ref', ascending=True)['date_ref'].dt.strftime('%Y-%m'), rotation=45)

ax.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: f'{int(x):,}'))

ax.set_xlabel('Date')
ax.set_ylabel('Value (R$)')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
df_billing.sort_values(by='date_ref', ascending=True).plot(kind="line", x='date_ref', y='value', legend=False, ax=ax)

ax.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: f'{int(x):,}'))

ax.set_xlabel('Date')
ax.set_ylabel('Value (R$)')

plt.tight_layout()
plt.show()
```
